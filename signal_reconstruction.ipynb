### Target: reconstruct the Input signal

Approach: to replace the real system problem by a problem involving mathematical relationships


Create a suitable model of the physical process and its constraints. Basic idea is to represent the gas analyser as
simple volumes with perfect gas mixing and mass diffusion. The assumption will be the gas passes through the volumes at
constant speed and the concentration gradient is in axial direction only. 



<div>
    <img src=img/gas_flow.jpg width=\"700\"/>
</div>



#### What is to be modeled phenomenologicaly:

* one dimensional mixing equation

\begin{equation*}
\lambda C_0(t)+\frac{dC(t)} {dt}=\lambda C_i(t)
\end{equation*}

where,  $\lambda $ - characteristic parameter (constant)

$C_0(t)$ -output concentration
$C_i(t)$ - input concentration


* Mass diffusion equation




\begin{equation*}
\frac{\partial C(x,t)}{\partial } = D \frac{\partial^2C(x,t)}{\partial ^2 } 
\end{equation*}
